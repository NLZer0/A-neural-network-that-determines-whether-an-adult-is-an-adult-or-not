```python
import os
```

```python
PATH = 'C:\\Users\\zanzi\\Desktop\\Python\\LAG_dataset\\data'
dataset = pd.DataFrame()
for name in os.listdir(PATH):
    items = np.asarray(os.listdir(PATH + '\\' + name))
    for young_img_label in os.listdir(PATH + '\\' + name + '\\y'):
        young_img_path = PATH + '\\' + name + '\\y' + '\\' + young_img_label
        dataset = dataset.append({'img_path': young_img_path, 'age': 'y'}, ignore_index=True)
    for adult_img_label in items[items!='y']:
        adult_img_path = PATH + '\\' + name + '\\' + adult_img_label
        dataset = dataset.append({'img_path': adult_img_path, 'age': 'a'}, ignore_index=True)
```

```python
dataset.to_csv('Parsed_data.csv', index=False)
```
